```python
import pandas as pd
import numpy as np
import numpy as np
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split

df = pd.read_csv('/content/drive/Shareddrives/CMPUT 664 : SSE/Data/adult census/adult.data',header= None,delimiter=",")
df[14] = np.where(df[14]==' <=50K', 0, 1)
df.columns = ['age','workclass','fnlwgt','education','education_num','marital_status','occupation','relationship','race','sex','capital_gain','capital_loss','hours_per_week','native_country','<50k']
df
```

```python
# Modifying continuous atatributes : age, fnlwgt,education_num, education-num,capital-gain,capital-loss, hours-per-week
import random
age = list(df.age.values)
fnlwgt = list(df.fnlwgt.values)
education_num = list(df.education_num.values)
capital_gain = list(df.capital_gain.values)
capital_loss = list(df.capital_loss.values)
hour_per_week = list(df.hours_per_week.values)


mod_age = []
mod_fnlget = []
mod_education_num = []
mod_capital_gain = []
mod_capital_loss = []
mod_hour_per_week = []

for i in range(len(age)):
  mod_age.append(age[i]* random.uniform(0.8, 1))
  mod_fnlget.append(fnlwgt[i]*random.uniform(0.8,1))
  
  mod_education_num.append(education_num[i]* random.uniform(0.8, 1))
  mod_capital_gain.append(capital_gain[i]* random.uniform(0.8, 1))
  mod_capital_loss.append(capital_loss[i]* random.uniform(0.8, 1))
  mod_hour_per_week.append(hour_per_week[i]* random.uniform(0.8, 1))  
```

```python
df['age'] = mod_age
df['fnlwgt'] = mod_fnlget
df['education_num'] = mod_education_num
df['capital_gain'] = mod_capital_gain
df['capital_loss'] = mod_capital_loss
df['hours_per_week'] = mod_hour_per_week
```

```python
df.to_csv('synthetic_data_adult_census.csv')
```
