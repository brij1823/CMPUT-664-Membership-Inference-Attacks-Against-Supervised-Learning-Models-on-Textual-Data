```python
import pandas as pd
import numpy as np
import numpy as np
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
import random
```

```python
df = pd.read_csv('diabetes.csv')
df.head()
```

```python
glucose = list(df['Glucose'].values)
bloodPressure = list(df['BloodPressure'].values)
skinThickness = list(df['SkinThickness'].values)
insulin = list(df['Insulin'].values)
bmi = list(df['BMI'].values)
diabetespedigreefunction = list(df['DiabetesPedigreeFunction'].values)


mod_glucose = []
mod_bp = []
mod_st = []
mod_insulin = []
mod_bmi = []
mod_diab = []

for i in range(len(glucose)):
  mod_glucose.append(glucose[i] * random.uniform(0.8, 1))
  mod_bp.append(bloodPressure[i]*random.betavariate(0.8,1))
  mod_st.append(skinThickness[i]* random.uniform(0.8, 1))
  mod_insulin.append(insulin[i]*random.uniform(0.8, 1))
  mod_bmi.append(bmi[i]*random.uniform(0.8, 1))
  mod_diab.append(diabetespedigreefunction[i]*random.uniform(0.8, 1))

```

```python
df.head()
```

```python
df['Glucose'] = mod_glucose
df['BloodPressure'] = mod_bp
df['SkinThickness'] = mod_st
df['BMI'] = mod_bmi
df['DiabetesPedigreeFunction'] = mod_diab
```

```python
df.to_csv('synthetic_pima_diabetic.csv')
```
