```python
import pandas as pd
import random
```

```python

df = pd.read_csv('cleveland.csv', header = None)

df.columns = ['age', 'sex', 'cp', 'trestbps', 'chol',
              'fbs', 'restecg', 'thalach', 'exang', 
              'oldpeak', 'slope', 'ca', 'thal', 'target']

```

```python
df.head()
```

```python
age = list(df.age.values)
trestbps = list(df.trestbps.values)
chol = list(df.chol.values)
thalach = list(df.thalach.values)
oldpeak = list(df.oldpeak.values)
thal = list(df.thal.values)

mod_age = []
mod_trestbps = []
mod_chol = []
mod_thalach = []
mod_oldpeak = []
mod_thal = []

for i in range(len(age)):
  mod_age.append(age[i]*random.uniform(0.8, 1))
  mod_trestbps.append(trestbps[i]*random.uniform(0.8, 1))
  mod_chol.append(chol[i]*random.uniform(0.8, 1))
  mod_thalach.append(thalach[i]*random.uniform(0.8, 1))
  mod_oldpeak.append(oldpeak[i]*random.uniform(0.8, 1))
  mod_thal.append(thal[i]*random.uniform(0.8, 1))
```

```python
df['trestbps'] = mod_trestbps
df['chol'] = mod_chol
df['thalach'] = mod_thalach
df['oldpeak'] = mod_oldpeak
df['thal'] = mod_thal
```

```python
df.head()
```

```python
df.to_csv('synthetic cleveland.csv')
```
